# ใบเกรด โดยใช้ภาษา Python

งานเก็บคะแนน ม.4

```python
def Grade(term = 1):
    subject = ['ภาษาไทย        ', 'คณิตศาสตร์ 1     ', 'คณิตศาสตร์ 2     ', 'ฟิสิกส์พื้นฐาน      ', 'เคมีพื้นฐาน       ', 'พื้นฐานของสิ่งมีชีวิต ', 'คอมพิวเตอร์เบื้องต้น', 'การออกแบบ      ', 'ศาสนศึกษา       ', 'สุขศึกษา         ', 'ทัศนศิลป์         ', 'ภาษาอังกฤษ      ']
    subjectcode = ['ท30101   ', 'ค30101   ', 'ค30102   ', 'ว30101   ', 'ว30121   ', 'ว30141   ', 'ว30181   ', 'ว30182   ' , 'ส30101   ', 'พ30101   ','ศ30101   ', 'อ30101   ']
    score = []
    credit = [1.0, 1.5, 2.0, 1.5, 1.5, 1.5, 1.0, 1.0, 1.0, 0.5, 0.5, 2.0]
    grade = []
    GPA = []
    name = str(input('ชื่อ: '))
    surname = str(input('นามสกุล: '))
    mathayom = int(input('มัธยมศึกษาปีที่: '))
    classroom = int(input('ห้อง: '))
    no = int(input('เลขที่: '))
    code = int(input('เลขประจำตัว: '))
    term = int(input('ภาคเรียนที่: ')) 
    year = int(input('ปีการศึกษา: '))

    while len(score)<len(subject):
        for x in range (len(subject)):
            point = float(input(subject[x]))
            score.append(point)
            if (point >= 80 and point <= 100):
                grade.append(4.00)  
            elif (point >= 75 and point < 80):
                grade.append(3.50)
            elif (point >= 70 and point < 75):
                grade.append(3.00)    
            elif (point < 70):
                grade.append(0.00) 
            else:
                print('มีข้อผิดพลาด โปรดใส่คะแนนใหม่')
            print('')
        for i in range (len(subject)):
            GPA.append(grade[i] * credit[i])
        a = float(sum(GPA)/sum(credit))
          
    print('-'*50)     
    print('')
    print('     ','โรงเรียนวิทยาศาสตร์จุฬาภรณราชวิทยาลัย ปทุมธานี')
    print('     '+'รายงานผลการเรียน', 'ภาคเรียนที่ ', term, 'ปีการศึกษา ', year)    
    print('ชื่อ', name, 'นามสกุล', surname,'มัธยมศึกษาปีที่', mathayom, 'ห้อง', classroom,'เลขที่', no, 'เลขประจำตัว', code)
    print('ครูที่ปรึกษา','   ', '1. นายสุเทพ ผานัด','   ', '2. นางสาวกานต์รพีณัชชาช์ ไชยปัญญา')
    print('')
    print('รหัสวิชา', '         ', 'วิชา','         ',    'หน่วยกิต', '   ', 'คะแนน', ' ', 'เกรด' )
    print('')
    
    for i in range (len(grade)):         
        print(subjectcode[i], subject[i],'     ', credit[i], '     ', score[i],'   ', grade[i])
    
    print('')
    print ('ผลการเรียนเฉลี่ย', '%.2f'%a)
    print('*หมายเหตุเกรด 0.0 คือ ร')
    print('')
    print('')
    print('')
    print('ลงชื่อ..........................................ครูที่ปรึกษา')
    print('(1. นายสุเทพ ผานัด         2. นางสาวกานต์รพีณัชชาช์ ไชยปัญญา)')
    print('')
    print('-'*50)
    print('')
    edit = str(input('พิมพ์ NextTerm เพื่อกรอกคะแนนเทอมต่อไป พิมพ์ Finish เพื่อจบการกรอกคะแนน: '))
    if edit == 'NextTerm':
        Grade(term + 1)
    else:
        print('เสร็จเรียบร้อย')
                         
Grade()
```
